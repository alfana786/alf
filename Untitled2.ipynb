# **3.Extraction of AMERICAN CUSTOMER SATISFICATION INDEX and Summarize Sentiment for Each Airline**

```python
import pandas as pd

# Step 1: Load Data
tweets_path = 'tweet.csv'  # Replace with your actual path
acsi_path = 'acsii.csv'  # Replace with your actual path

# Load datasets
tweets_data = pd.read_csv(tweets_path)

# Create a DataFrame for  ACSI data
acsi_data = pd.DataFrame({
    'Airline': ['JetBlue', 'American', 'Delta', 'Southwest', 'United', 'Alaska',
                'All Others', 'Allegiant', 'Frontier', 'Spirit'],
    '2021': [76, 77, 75, 79, 79, 75, 77, 74, 72, 68],
    '2022': [75, 79, 77, 77, 77, 77, 75, 71, 70, 66],
    '% Change': ['-1.3%', '+3%', '+3%', '-3%', '-3%', '+3%', '-3%', '-4%', '-3%', '-3%']
})

# Step 2: Define Sentiment Keywords
positive_keywords = ['great', 'excellent', 'amazing', 'good', 'happy', 'satisfied', 'love']
negative_keywords =  ['terrible', 'horrible', 'Late Flight','worst', 'canttell','angry', 'Cancelled Flight','Damaged Luggage','Bad Flight', 'hate']

# Step 3: Sentiment Analysis on Tweets
def evaluate_sentiment(text):
    """Evaluate sentiment based on presence of keywords."""
    text = str(text).lower()
    if any(word in text for word in positive_keywords):
        return 'positive'
    elif any(word in text for word in negative_keywords):
        return 'negative'
    else:
        return 'neutral'

# Apply sentiment analysis
tweets_data['evaluated_sentiment'] = tweets_data['text'].apply(evaluate_sentiment)

# 3 Step 4: Summarize Sentiment for Each Airline
sentiment_summary = tweets_data.groupby(['airline', 'evaluated_sentiment']).size().unstack(fill_value=0)

# Step 5: Calculate Twitter Sentiment Scores
sentiment_summary['twitter_score'] = (sentiment_summary.get('positive', 0) - sentiment_summary.get('negative', 0)) / sentiment_summary.sum(axis=1)

# Step 6: Merge Sentiment with Updated ACSI Data
# Merge based on airline names
comparison_df = sentiment_summary[['twitter_score']].merge(
    acsi_data.set_index('Airline'),
    left_index=True,
    right_index=True,
    how='inner'
)

# Output Results
print("Sentiment Summary by Airline:")
print(sentiment_summary)
print("\nUpdated ACSI Data:")
print(acsi_data)
print("\nComparison of Twitter Sentiment with ACSI Scores:")
print(comparison_df)
```

# **5.COMPARISION FROM TWITTER SCORE WITH ACSI**





```python
import matplotlib.pyplot as plt
import numpy as np

# Prepare data
airlines = comparison_df.index.tolist()
acsi_2021 = comparison_df['2021']
acsi_2022 = comparison_df['2022']
twitter_scores = comparison_df['twitter_score'] * 100  # Adjusted scaling

# Set bar positions and bar height
bar_height = 0.2
y_positions = np.arange(len(airlines))

# Create the plot
plt.figure(figsize=(14, 8))

# Plot horizontal bars
bars_2021 = plt.barh(y_positions - bar_height, acsi_2021, bar_height, color='blue', label='ACSI 2021')
bars_2022 = plt.barh(y_positions, acsi_2022, bar_height, color='orange', label='ACSI 2022')
bars_twitter = plt.barh(y_positions + bar_height, twitter_scores, bar_height, color='green', label='Twitter Sentiment Score')

# Add labels for each bar
for bar in bars_2021:
    plt.text(bar.get_width() + 0.5, bar.get_y() + bar.get_height() / 2,
             f'{bar.get_width():.1f}', va='center', fontsize=9, color='blue')
for bar in bars_2022:
    plt.text(bar.get_width() + 0.5, bar.get_y() + bar.get_height() / 2,
             f'{bar.get_width():.1f}', va='center', fontsize=9, color='orange')
for bar in bars_twitter:
    plt.text(bar.get_width() + 0.5, bar.get_y() + bar.get_height() / 2,
             f'{bar.get_width():.1f}', va='center', fontsize=9, color='green')

# Add axis labels, title, and legend
plt.yticks(y_positions, airlines)
plt.xlabel('Score')
plt.title('ACSI Scores and Twitter Sentiment by Airline')
plt.axvline(50, color='gray', linestyle='--', linewidth=0.8, alpha=0.7)  # Optional reference line
plt.legend()

# Add grid for readability
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Ensure proper spacing
plt.tight_layout()

# Show the plot
plt.show()
```

**TOTAL TWEETS PER AIRLINE**

```python
import matplotlib.pyplot as plt

# Step 1: Count the total number of tweets per airline
tweet_counts = tweets_data['airline'].value_counts()

# Step 2: Calculate the percentage of total tweets for each airline
total_tweets = tweet_counts.sum()
tweet_percentage = (tweet_counts / total_tweets) * 100

# Step 3: Plot the total number of tweets for each airline
ax = tweet_counts.plot(kind='bar', figsize=(10, 6), color='skyblue')

# Add labels and title
plt.xlabel('Airline')
plt.ylabel('Number of Tweets')
plt.title('Total Number of Tweets per Airline (in percentage)')

# Add percentage labels for each bar
for i, percentage in enumerate(tweet_percentage):
    ax.text(i, tweet_counts[i] + 5, f'{percentage:.1f}%', ha='center', fontsize=10, color='black')

# Show the plot
plt.tight_layout()
plt.show()
```
